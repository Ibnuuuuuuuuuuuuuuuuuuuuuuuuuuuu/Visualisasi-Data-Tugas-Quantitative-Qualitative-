```python
uploaded = files.upload()

uploaded_filename = list(uploaded.keys())[0]

df = pd.read_excel(io.BytesIO(uploaded[uploaded_filename]))
```

```python
col_price = 'unit_price'
N_BINS_PRICE = 4
KATEGORI_LABEL_HARGA = ['Rendah', 'Sedang', 'Tinggi', 'Sangat Tinggi']

print(f"\nLangkah 3a: Diskretisasi '{col_price}' menggunakan Frekuensi-Sama (pd.qcut) ({N_BINS_PRICE} Bins)...")

try:
    df[f'{col_price}_Kategori'] = pd.qcut(
        df[col_price],
        q=N_BINS_PRICE,
        labels=KATEGORI_LABEL_HARGA,
        duplicates='drop'
    )
    print(f"  - Kolom '{col_price}' berhasil diubah ke '{col_price}_Kategori'.")
    print(f"    > Distribusi Kategori {col_price} (Frekuensi-Sama):")
    print(df[f'{col_price}_Kategori'].value_counts().sort_index())

except ValueError as e:
    print(f"  - ERROR pada kolom {col_price}. Diskretisasi dibatalkan. Pesan: {e}")

col_qty = 'transaction_qty'
QTY_MAPPING = {1: 'Satuan', 2: 'Dua'}

print(f"\nLangkah 3b: Mengategorikan '{col_qty}' menggunakan Pemetaan Langsung...")

df[f'{col_qty}_Kategori'] = df[col_qty].map(QTY_MAPPING)

print(f"  - Kolom '{col_qty}' berhasil diubah ke '{col_qty}_Kategori'.")
print(f"    > Distribusi Kategori {col_qty} (Pemetaan Langsung):")
print(df[f'{col_qty}_Kategori'].value_counts())

print("\nLangkah 4: DataFrame Akhir (Kolom Kategorikal Berupa Teks):")

display_cols = ['transaction_id', 'transaction_qty', 'unit_price',
                'unit_price_Kategori', 'transaction_qty_Kategori']

valid_display_cols = [c for c in display_cols if c in df.columns]

print(df[valid_display_cols].head(10))

print("\n--- SKRIP SELESAI ---")
```
